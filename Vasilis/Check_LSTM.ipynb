```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
model_values = pd.read_csv('inference_full_results_00.csv',delimiter=';')
```

```python
model_values.head(20)
```

```python
# Train dataframe: keep only train columns
train_df = model_values.drop(
    columns=["timestamp_validation","predicted_delay_validation", "true_delay_validation", "timestamp_test","predicted_delay_test", "true_delay_test"]
).dropna()

# Validation dataframe: keep only validation columns
val_df = model_values.drop(
    columns=["timestamp_train","predicted_delay_train", "true_delay_train", "timestamp_test","predicted_delay_test", "true_delay_test"]
).dropna()

# Test dataframe: keep only test columns
test_df = model_values.drop(
    columns=["timestamp_train","predicted_delay_train", "true_delay_train", "timestamp_validation","predicted_delay_validation", "true_delay_validation"]
).dropna()
```

```python
train_df.shape
```

```python
train_df = train_df.copy()
train_df["residuals"] = train_df["true_delay_train"] - train_df["predicted_delay_train"]

val_df = val_df.copy()
val_df["residuals"] = val_df["true_delay_validation"] - val_df["predicted_delay_validation"]

test_df = test_df.copy()
test_df["residuals"] = test_df["true_delay_test"] - test_df["predicted_delay_test"]

```

```python
train_df.head()
```

```python
train_df = train_df.rename(columns={
    "timestamp_train": "timestamp",
    "true_delay_train": "ytrue",
    "predicted_delay_train": "ypred"
})

val_df = val_df.rename(columns={
    "timestamp_validation": "timestamp",
    "true_delay_validation": "ytrue",
    "predicted_delay_validation": "ypred"
})

test_df = test_df.rename(columns={
    "timestamp_test": "timestamp",
    "true_delay_test": "ytrue",
    "predicted_delay_test": "ypred"
})
```

```python
train_df.head()
```

```python
train_df.to_csv("lstm_train_predictions.csv")
val_df.to_csv("lstm_val_predictions.csv")
test_df.to_csv("lstm_test_predictions.csv")

print("Saved successfully!")
```
