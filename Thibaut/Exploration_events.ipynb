```python
import pandas as pd
```

```python
df = pd.read_csv('ABA Datasets/nyc_events_filtered.csv')
```

```python
df.head(5)
```

```python
# count the number of entries per police precinct
precinct_counts = df['Police Precinct'].value_counts()
# order the counts by value count and not by index
precinct_counts = precinct_counts.sort_values(ascending=False)
# print the counts
for precinct, count in precinct_counts.items():
    print(f'Police Precinct: {precinct}, Number of Entries: {count}')
```

```python
# Check for a specific month, e.g. August
month = '12'
df['Month'] = pd.to_datetime(df['Start Date/Time']).dt.month.astype(str).str.zfill(2)
df_august = df[df['Month'] == month]
print(f'Number of entries in August: {len(df_august)}')

# count the number of entries per police precinct
precinct_counts = df_august['Police Precinct'].value_counts()
# order the counts by value count and not by index
precinct_counts = precinct_counts.sort_values(ascending=False)
# print the counts
for precinct, count in precinct_counts.items():
    print(f'Police Precinct: {precinct}, Number of Entries: {count}')
```
